```python
import numpy as np

from matplotlib import pyplot as plt

from sklearn import manifold
```

```python
similarities = np.genfromtxt('RuleSimilarityDistance.csv', delimiter=',')
```

```python
EPSILON = np.finfo(np.float32).eps
seed = np.random.RandomState(seed=3)

mds = manifold.MDS(
    n_components=2,
    max_iter=3000,
    eps=1e-9,
    random_state=seed,
    dissimilarity="precomputed",
    n_jobs=1,
    normalized_stress="auto",
)
pos = mds.fit(similarities).embedding_

fig = plt.figure(1)
ax = plt.axes([0.0, 0.0, 1.0, 1.0])

s = 100
plt.scatter(pos[:, 0], pos[:, 1], color="turquoise", s=s, lw=0, label="MDS")
plt.legend(scatterpoints=1, loc="best", shadow=False)

similarities = similarities.max() / (similarities + EPSILON) * 100
np.fill_diagonal(similarities, 0)

for i in range(len(pos)):
    ax.annotate(i, (pos[i,0], pos[i,1]))

plt.show()
```

```python
def is_metric(distance_matrix: np.ndarray):
    D = distance_matrix
    N = distance_matrix.shape[0]

    for i in range(N):
        for j in range(i):
            if not (D[i, j] <= D[i, :] + D[:, j]).all():
                print("i is {} and j is {}".format(i,j))
                return False

    return True
```

```python
is_metric(similarities)
```
